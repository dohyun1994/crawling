#### reshape
- 차원과 크기를 변경
- 변경되는 전체의 수가 변경 전의 전체수가 동일해야 함 
- ndim : 차원수

```python
arr = np.arange(12)
print(arr)
print(arr.shape)
print(arr.ndim)
```

```python
arr2 = arr.reshape(3, 4)
print(arr2)
print(arr2.shape)
print(arr2.ndim)
```

```python
arr3 = arr.reshape(2, 3, 2)
print(arr3)
print(arr3.shape)
print(arr3.ndim)
```

```python
arr4 = arr.reshape(2, 3, 2)
print(arr4)
print(arr4.shape)
print(arr4.ndim)
```

```python
arr5 = arr4.reshape(12)
print(arr5)
print(arr5.shape)
print(arr5.ndim)
```

#### 다차원 배열을 1차원으로
- ravel
- flatten
- reshape
- flatten은 값을 복사해서 새로운 배열을 생성하고 ravel, reshape는 값을 참조하여 생성
    - https://m.blog.naver.com/wideeyed/221533365486

```python
arr1 = np.arange(12).reshape(3, 4)
print(arr1)
```

```python
arr2 = arr1.ravel()
print(arr2)
```

```python
arr3 = arr1.reshape(-1)
print(arr3)
```

```python
arr4 = arr1.flatten()
print(arr4)
```
