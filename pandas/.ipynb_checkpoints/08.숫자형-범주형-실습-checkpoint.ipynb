```python
import math
```

```python
def age_cate(age):
    if math.isnan(age):
        return "나이미상"
    else:
        firstage = str(int(age/10))
        if firstage == '0':
            return '10대미만'
        else:
            return firstage + '0대'
        
```

```python
for i in np.arange(df['Age'].shape[0]):
    #print(df.loc[i, 'Age'])
    print(age_cate(df.loc[i, 'Age']))
```

```python
df['연령대'] = df['Age'].apply(age_cate)
df.head()
```

```python
df[df['Age'].isna()]
```

```python
def hap(x, y):
    return x + y

hap(10, 20)
```

```python
def sex_cate(sex):
    if sex == 'male':
        return '남자'
    else:
        return '여자'
       
```

```python
sex_cate('male')
```

```python
df['성별'] = df['Sex'].apply(sex_cate)
df.head()
```

```python
def survived_pclass(p_df):
    str(p_df['Pclass']) + ':' + ('생존' if p_df['Survived'] == 1 else '사망')
```

```python
df['Sur2'] = df.apply(survived_pclass, axis=1)
```
