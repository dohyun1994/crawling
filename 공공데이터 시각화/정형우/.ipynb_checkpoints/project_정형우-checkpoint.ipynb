```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
```

```python
# csv파일 불러오기(한글깨짐 현상 해결)
df = pd.read_csv('./fire_2019_eng.csv',sep=',',encoding='CP949')
df
```

```python
# 일시를 datetime형으로 변경
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d %H:%M:%S', errors='raise')
df.info()
```

```python
# 화재발생일시의 값을 "년, 월, 일, 시, 분, 요일"로 분리하여 데이터 추가
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
df['day'] = df['Date'].dt.day
df['hour'] = df['Date'].dt.hour
df['minute'] = df['Date'].dt.minute
df['weekday'] = df['Date'].dt.weekday
df
```

```python
plt.rc('font', family='Malgun Gothic') # 그래프 한글깨짐 해결
plt.rcParams['axes.grid'] = False # 눈금표시 제거

# [S] 시각화 할 데이터 추출
gb1 = df.groupby(by='month')

fireData = gb1[['Date']].count()
fireData = fireData.rename(columns={'Date':'화재건수'})
fireData['총피해금액'] = gb1[['Moneys']].sum()
fireData['평균피해금액'] = gb1[['Moneys']].mean().round()
fireData = fireData.astype({'평균피해금액':'int'})
fireData = fireData.reset_index()
fireData = fireData.rename(columns={'month':'월'})
# [E] 시각화 할 데이터 추출

# [S] 데이터 시각화
x = fireData['월']
y1 = fireData['화재건수']
y2 = fireData['총피해금액']
y3 = fireData['평균피해금액']

fig = plt.figure(figsize=(10,6)) ## 캔버스 생성
fig.set_facecolor('white')
ax1 = fig.add_subplot() ## axes 생성

config_plot = dict( ## 키워드 인자
    linestyle='solid', # 선 스타일
    linewidth=2, # 선 두께 
    marker='o', # 마커 모양
    markersize=10, # 마커 사이즈
    markeredgecolor='white', # 마커 테두리 색깔
    markeredgewidth=3, # 마커 테두리 두께
)

color1 = '#b6b6b6' # 색상설정
ax1.bar(x, y1, color=color1) # x,y축 데이터 및 색상값지정
ax1.set_xlabel('월') # X축 label변경
ax1.set_ylabel('화재건수', color=color1) # Y축 label명 및 그래프색상변경
ax1.tick_params(axis='y', labelcolor=color1) # Y축 label색상변경
ax1.grid(True, axis='x') # X축 눈금만 사용

color2 = '#ff6864'
ax2 = ax1.twinx() # 위에 선언한 ax1의 x축 사용
ax2.plot(x, y2, color=color2,**config_plot) # x,y축 데이터 및 색상값지정, config_plot를 이용하여 마커생성
ax2.set_ylabel('총피해금액', color=color2)
ax2.tick_params(axis='y', labelcolor=color2)
 
color3 = '#598686'
ax3 = ax1.twinx()
ax3.plot(x, y3, color=color3,**config_plot)
ax3.set_ylabel('평균피해금액', color=color3)
ax3.tick_params(axis='y', labelcolor=color3)
ax3.spines.right.set_position(("axes", 1.2)) ## 오른쪽 옆에 y축 추가

plt.title('2019년 월별 화재발생건수 및 피해액',fontsize=15) # 타이틀 설정
plt.show() # 그래프 출력

display(fireData) # 시각화에 사용한 데이터 출력
# [E] 데이터 시각화
```

```python
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.grid'] = False 
korWeek = ['월요일', '화요일', '수요일', '목요일', '금요일', '토요일', '일요일']
def kweek(x) :
  return korWeek[x]

gb1 = df.groupby(by='weekday')

fireData = gb1[['Date']].count()
fireData = fireData.rename(columns={'Date':'화재건수'})
fireData['총피해금액'] = gb1[['Moneys']].sum()
fireData['평균피해금액'] = gb1[['Moneys']].mean().round()
fireData = fireData.astype({'평균피해금액':'int'})
fireData = fireData.reset_index()
fireData['weekday'] = fireData['weekday'].apply(kweek)
fireData = fireData.rename(columns={'weekday':'요일'})

x = fireData['요일']
y1 = fireData['화재건수']
y2 = fireData['총피해금액']
y3 = fireData['평균피해금액']

fig = plt.figure(figsize=(13,6)) ## 캔버스 생성
fig.set_facecolor('white')
ax1 = fig.add_subplot() ## axes 생성

config_plot = dict( ## 키워드 인자
    linestyle='solid', # 선 스타일
    linewidth=2, # 선 두께 
    marker='o', # 마커 모양
    markersize=10, # 마커 사이즈
    markeredgecolor='white', # 마커 테두리 색깔
    markeredgewidth=3, # 마커 테두리 두께
)

color1 = '#b6b6b6'
ax1.bar(x, y1, color=color1)
ax1.set_xlabel('요일')
ax1.set_ylabel('화재건수', color=color1)
ax1.tick_params(axis='y', labelcolor=color1)
ax1.grid(True, axis='x')

color2 = '#ff6864'
ax2 = ax1.twinx()
ax2.plot(x, y2, color=color2,**config_plot)
ax2.set_ylabel('총피해금액', color=color2)
ax2.tick_params(axis='y', labelcolor=color2)
 
color3 = '#598686'
ax3 = ax1.twinx()
ax3.plot(x, y3, color=color3,**config_plot)
ax3.set_ylabel('평균피해금액', color=color3)
ax3.tick_params(axis='y', labelcolor=color3)
ax3.spines.right.set_position(("axes", 1.2)) ## 오른쪽 옆에 y축 추가

plt.title('2019년 요일별 화재발생건수 및 피해액',fontsize=15)
plt.show()

display(fireData)
```
