```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
```

```python
# csv파일 불러오기(한글깨짐 현상 해결)
df = pd.read_csv('./fire_2019_eng.csv',sep=',',encoding='CP949')
df
```

```python
# 일시를 datetime형으로 변경
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d %H:%M:%S', errors='raise')
df.info()
```

```python
# 화재발생일시의 값을 "년, 월, 일, 시, 분, 요일"로 분리하여 데이터 추가
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
df['day'] = df['Date'].dt.day
df['hour'] = df['Date'].dt.hour
df['minute'] = df['Date'].dt.minute
df['weekday'] = df['Date'].dt.weekday
df
```

```python
plt.rc('font', family='Malgun Gothic') # 그래프 한글깨짐 해결
plt.rcParams['axes.grid'] = False # 눈금표시 제거

# [S] 시각화 할 데이터 추출
gb1 = df.groupby(by='month')

fireData = gb1[['Date']].count()
fireData
```
