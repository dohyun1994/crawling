```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./scv.csv',encoding='cp949')

df
```

```python
df2 = pd.read_csv('./scv.csv', \
                  usecols=['시도','재산피해소계'],encoding='cp949')
df2
```

```python
gb1 = df2.groupby(by='시도')
gb1_gb=gb1.sum().sort_values(by='재산피해소계' ,ascending=True)
gb1_gb=gb1_gb.reset_index()
gb1_gb
```

```python
gb1_gb.plot('시도', '재산피해소계', kind='barh')
```
