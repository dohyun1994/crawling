```python
import numpy as np
```

## 인덱싱
- 단일 값: 인덱스를 이용하여 하나의 데이터에 접근
- 슬라이싱: 연속된 인덱스를 이용하여 여러 데이터에 접근
- 팬시 인덱싱 : 리스트 형태의 인덱스를 이ㅛㅇㅇ하여 여러 데이터에 접근
- 불린 인덱싱 : 불린값의 리스트 형태의 인덱스를 이용하여 여러 데이터에 접근

```python
arr = np.arange(1,11)
print(arr)
```

### 2차원


```python
arr = np.arange(1,10).reshape(3, -1)
print(arr)
```

```python
print(arr[0])
```

```python
print(arr[0].shape)
print(arr[0].ndim)
```

```python
print(arr[2,1])
```

```python
print(arr[0].shape)
print(arr[0].ndim)
```

```python
print(arr[0, -1])
```

## 3차원

```python
arr = np.arange(1,9).reshape(2,2,2)
print(arr)
print(arr.shape)
```

```python
print(arr[0])
```

```python
print(arr[0])
print("")
print(arr[0,0])
print("")
print(arr[0].ndim)
```

```python
print(arr[1,1,1])
```

## 슬라이싱
- 1차원

```python
arr = np.arange(1, 11)
print(arr)
```

```python
print(arr[1:5])
```

```python
print(arr[2:])
```

```python
print(arr[:8])
```

```python
print(arr[-2:])
```

- 2차원

```python
arr = np.arange(1,10).reshape(3, -1)
print(arr)
```

```python
print(arr[0:2,1:2])
print(arr[0:2,1:2].shape)
print(arr[0:2,1:2].ndim)
```

### 4,5 / 7,8 가져오기

```python
print(arr[1:3,0:2])   # 1~2행 슬라이싱 후, 1~2열 슬라이싱
```

```python
print(arr[1:,0:2])   # 위와 동일 결과.
```

```python
print(arr[2, 0:2])
```

```python
arr = np.arange(16).reshape(4, -1)
print(arr)
```

### 10,11,14,15 가져오기

```python
print(arr[2:, 2:])
```

### 9,10,11 가져오기

```python
print(arr[2:3, 1:])
```

- 3차원

```python
arr = np.arange(27).reshape(3,3,3)
print(arr)
```

```python
print(arr[0:2, 0:2, 0:2])
```

```python
print(arr[:, :, :])  # 모든 배열 원소 다가져오기
```

### 펜시인덱싱
- 1차원

```python
arr = np.arange(1,11)
print(arr)
```

```python
arr = np.arange(1, 10).reshape(3, -1)
print(arr)
```

```python
print(arr[[0,2],2])
```

## 불린 인덱싱

```python
arr = np.arange(1, 11)
print(arr)
```

```python
arr > 3
```

```python
mask_1 = arr >3
print(mask_1)
```

```python
arr[[False, False, False,  True,  True,  True,  True,  True,  True,  True]]
```

```python
arr[mask_1]
```

```python
arr[arr > 3]
```

## 여러조건
    - and : &
    - or : |

```python
mask2 = (arr > 3) & (arr < 7)
print(mask2)
```

```python
arr[mask2]
```

```python
arr[ (arr>3) & (arr < 7)]
```
