- astype

```python
arr = np.array( [1, 2, 3])
print(arr.dtype)

arr2 = arr.astype('float64')
print(arr2)
print(arr2.dtype)
```

```python
arr = np.array( ["1", "2", "3"])
print(arr.dtype)

arr2 = arr.astype('int32')
print(arr2)
print(arr2.dtype)
```

#### 여러가지 생성 방법.

- arange

```python
for i in range(1,10):
    print(i)
```

```python
type(np.arange(1,10))
```

```python
arr = np.array(np.arange(1,10))
print(arr)
```

- ones

```python
arr = np.ones((2,3,4), dtype=int)
print(arr)
```

- zeros

```python
arr = np.zeros((2,3), dtype=int)
print(arr)
```

```python
arr = np.zeros((2,3,4), dtype=int)
print(arr)
```

- empty

```python
arr = np.empty((2,3,5))
print(arr)
```

- full

```python
arr = np.full((2,3), 4, dtype=int)
print(arr)
```

```python
array = np.full((5,7), 3)
print(array)
```

```python
arr = np.full((5,2), [4, 5], dtype=int)
print(arr)
```
