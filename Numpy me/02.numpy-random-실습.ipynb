```python
# numpy 사용
import numpy as np
```

```python
import random

# 0과 1미만 사이의 랜덤값
print(random.random())

# 범위(1이상 10 미만 중에 2간격으로 random값 - 정수값 return)
rn1 = random.randrange(1,10,2)
print(rn1)

# 범위(1이상 10이하 - 정수값 return)
rn2 = random.randint(1,10)
print(rn2)
```

### rand : 0과 1 사이의 랜덤값(균등분포)

```python
arr = np.random.rand(2, 3, 4)
print(arr)
```

### randn : 정규분포, 가우시안 분포 난수값

```python
arr = np.random.randn(2, 3, 4)
print(arr)
```

#### randint: 정수 난수 값

```python
# 0이상 3미만의 랜덤값.
arr = np.random.randint(3, size=6)
print(arr)
```

```python
# 3이상 10미만의 랜덤값.
arr = np.random.randint(3,10, size=6)
print(arr)
```

```python
# 3이상 10미만의 랜덤값.   size 2X4
arr = np.random.randint(3,10, size=(3,4))
print(arr)
```

```python
np.random.seed(35)
arr = np.random.randn(2, 3)
print(arr)
```
