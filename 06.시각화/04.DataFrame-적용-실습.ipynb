### DataFrame를 바로 적용하여 시각화

```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].value_counts()
```

- line

```python
iris.info()
```

```python
iris['sepal_length'].plot(kind='line')  # x축 : index / y축 : iris['sepal_length']
```

```python
iris.plot(kind='line')
```

```python
iris.groupby('species').mean().reset_index()
```

```python
iris_gb = iris.groupby('species').mean().reset_index()
iris_gb
```

```python
iris_gb.plot('species', 'sepal_length', kind='line')  # (x, y, kind)
```

```python
iris.groupby('species').mean().reset_index().plot('species', 'sepal_length', kind='line')  # (x, y, kind)
```
