```python
from bs4 import BeautifulSoup
```

```python
html_doc = """<html><head><title>The Dormouse's story</title></head>
<body>
<p class="title"><b>The Dormouse's story</b></p>

<p class="story">Once upon a time there were three little sisters; and their names were
    <a href="http://example.com/elsie" class="sister" id="link1">Elsie</a>,
    <a href="http://example.com/lacie" class="sister" id="link2" data01="title_01">Lacie</a>
     and
    <a href="http://example.com/tillie" class="sister" id="link3" data02="title_02">Tillie</a>;
    and they lived at the bottom of a well.
</p>

<p class="story">...</p>
</body>
</html>
"""
```

```python
soup=BeautifulSoup(html_doc, 'html.parser')
soup
```

```python
print(soup.prettify())
```

```python
# <title>
soup.title
```

```python
soup.title.string
```

```python
soup.title.text
```

```python
soup.title.parent.name
```

```python
soup.p['class']
```

## find_all(), find()
- 태그를 이용하여 검색

## select().select_one()
- css선택자를 이용하여 검색

```python
html2 = soup.find('p')
html2
```

```python
html2 = soup.find('p', class_ = 'title')
html2
```

```python
html2 = soup.find('p', class_ = 'story')
html2
```

```python
html2 = soup.find('p', 'title')
html2
```

```python
html2 = soup.find('a', id='link1')
html2
```

```python
html2 = soup.find('a', attrs={'data01':'title_01'})
html2
```

```python
html2 = soup.find('a', attrs={'class':'sister', 'data01':'title_01'})
html2
```

```python
html2 = soup.find_all('p')
html2
```

```python
type(html2)
```

```python
for html in html2:
    print(html)
    print('---------------------------------')
```

```python
html2 = soup.find_all('p', class_='title')
html2
```

```python
for html in html2:
    print(html.text)
```

```python
html2 =soup.find_all('p', class_='story')
for html in html2:
    print(html.text)
    print('-------------------')
```
