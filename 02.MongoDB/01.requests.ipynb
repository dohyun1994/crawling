라이브러리 : requests

http request -> 웹을 통해 requests 요청

```python
import requests
```

```python
req = requests.get('https://pythonscraping.com/pages/page1.html')
req
```

```python
type(req)
```

```python
req_html_text = req.text
req_html_text
```

```python
req_header = req.headers
print(req_header)
print(type(req_header))
```

```python
req_header['content-type']
```

```python
req_status = req.status_code
req_status
```

- 다른예제


```python
r = requests.get('https://api.github.com')
print(r.status_code)
print(r.headers['content-type'])
print(r.encoding)
print(r.text)
print(r.json())
```

- https://httpbin.org
- https://httpbin.org/get?key1=val&key2=val2

```python
param_data = {'key1':'val1', 'key2':'val2'}
r = requests.get('http://httpbin.org/get', params = param_data)
r.text
```

```python
r.url
```

```python
import json
json_obj = json.loads(r.text)
json_obj
```

```python
r = requests.post('https://httpbin.org/post', data = {'key': 'value'})
r.text
```

```python
json_obj = json.loads(r.text)
json_obj
```

```python
param_data = {'key1' : 'val1', 'key2':['val2', 'val3']}
r = requests.post('http://httpbin.org/get', data = param_data)
r.request.body
```
