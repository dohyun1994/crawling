- 조건

```python
# where name = '홍길동'
srs = collection.find({'name' : '홍길동'})
for sr in srs:
    print(sr)
```

```python
# where name = '홍길동' and age = 17
srs = collection.find({'name': '홍길동', 'age': 17})
for sr in srs:
    print(sr)
```

```python
# where age >= 17
srs = collection.find({ 'age': {'$gte':17} })
for sr in srs:
    print(sr)
```

```python
# select name from collection where and age >= 17
srs = collection.find({'age' : {'$gte': 17}}, {'name':1, '_id':0 })
for sr in srs:
    print(sr)
```

- update_one, update_many

```python
# update
collection.update_one( {'name': '홍길동'}, {'$set': {'age': 18}})
```

```python
srs = collection.find({'name': '홍길동'})
for sr in srs:
    print(sr)
```

```python
# 나이가 30보다 많은 사람의 나이를 31로 update 하세요
collection.update_many( { 'age': {'$gt': 30} }, {'$set': {'age': 31} } )
```

```python
srs = collection.find()
for sr in srs:
    print(sr)
```

#### delete, delete_many

```python
collection.delete_one( {'name': '홍길동'})
```

```python
srs = collection.find()
for sr in srs:
    print(sr)
```

```python
collection.delete_many( {'age': {'$gte': 31}})
```

```python
srs = collection.find()
for sr in srs:
    print(sr)
```
